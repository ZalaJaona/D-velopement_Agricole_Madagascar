```python
import matplotlib.pyplot as plt

# Données
secteurs = [
    "Scolarisation primaire",
    "Terres titrées",
    "Semences certifiées",
    "Sécurité des fermes",
    "Transformation agro",
    "Agriculture dans le PIB"
]

taux_actuels = [56.9, 10, 9, 10, 10, 21.3]
taux_projetes = [100, 80, 50, 100, 100, 30]

# Paramètres du graphique
bar_width = 0.35
index = range(len(secteurs))

plt.figure(figsize=(10, 6))

# Barres
bars1 = plt.bar(index, taux_actuels, bar_width, label='Actuel', color='skyblue')
bars2 = plt.bar([i + bar_width for i in index], taux_projetes, bar_width, label='Projeté', color='seagreen')

# Ajout des pourcentages sur les barres
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

# Personnalisation
plt.xlabel('Secteurs')
plt.ylabel('Taux (%)')
plt.title('Taux par secteur : Actuel vs Projeté')
plt.xticks([i + bar_width / 2 for i in index], secteurs, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Catégories animales
categories = ['Bovins (têtes)', 'Caprins (têtes)', 'Œufs (unités)', 'Lait (litres)']

# Données actuelles et projetées
actuel = [9_100_000, 1_200_000, 465_000_000, 100_000]
objectif = [12_000_000, 2_000_000, 700_000_000, 300_000]

# Conversion en millions pour homogénéité visuelle
actuel_m = [val / 1_000_000 for val in actuel]
objectif_m = [val / 1_000_000 for val in objectif]

# Largeur des barres
bar_width = 0.35
index = range(len(categories))

# Création du graphique
plt.figure(figsize=(10, 6))
bars1 = plt.bar(index, actuel_m, bar_width, label='Actuel', color='sienna')
bars2 = plt.bar([i + bar_width for i in index], objectif_m, bar_width, label='Objectif', color='darkgreen')

# Ajout des annotations
for i in range(len(categories)):
    plt.text(i, actuel_m[i] + 0.1, f'{actuel[i]:,}', ha='center', va='bottom', fontsize=9)
    plt.text(i + bar_width, objectif_m[i] + 0.1, f'{objectif[i]:,}', ha='center', va='bottom', fontsize=9)

# Personnalisation
plt.xlabel('Catégories animales')
plt.ylabel('Volume (en millions ou milliers)')
plt.title('Cheptel et production animale – Actuel vs Objectif')
plt.xticks([i + bar_width / 2 for i in index], categories, rotation=15)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
import pdfplumber

chemin_pdf = r"C:\Users\zala_\Desktop\MADA\INSTAT_RGPH3-Definitif-ResultatsGlogaux-Tome1_17-2021.pdf"

regions = []

with pdfplumber.open(chemin_pdf) as pdf:
    for page_num in [148, 149, 150]:  # pages sont indexées à partir de 0
        texte = pdf.pages[page_num].extract_text()
        for ligne in texte.split('\n'):
            ligne = ligne.strip()
            if ligne and not ligne.isdigit() and len(ligne.split()) <= 4:
                regions.append(ligne)

print("Régions extraites :", regions)
```
